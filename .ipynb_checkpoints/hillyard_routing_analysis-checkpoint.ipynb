```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
file_path = "C:/Users/micha/OneDrive/Documents/GitHub/hillyard_routing_analysis/Resources/routing_analysis.csv"
data = pd.read_csv(file_path)
```

```python
selected_columns = ['Latitude', 'Longitude', 'Total Orders', 'Total Deliveries', 'Avg. Revenue']
X = data[selected_columns]
```

```python
X.fillna(0, inplace=True)
```

```python
k = 40  # Example: Choosing 3 clusters
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')
plt.title('K-means Clustering')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```
